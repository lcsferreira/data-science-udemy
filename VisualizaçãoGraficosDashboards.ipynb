## Gráfico de dispersão com legendas

```python
# Carregamento da base de dados
base = pd.read_csv('co2.csv')
base.head()
```

```python
#criando duas variáveis para cada atributo (x = conc e y = uptake)
x = base.conc
y = base.uptake
```

```python
# Retorna os valores únicos do atributo "treatment"
unicos = list(set(base.Treatment))
unicos
```

```python
# Percorre cada tipo de tratamento (chilled e nonchilled) e cria um gráfico de dispersão
for i in range(len(unicos)):
    indice = base.Treatment == unicos[i]
    plt.scatter(x[indice], y[indice], label = unicos[i])
plt.legend(loc = 'lower right')
```

## Gráfico de Dispersão com Seaborn

```python
# Importação das bibliotecas
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados
base = pd.read_csv('co2.csv')
base.head()
```

```python
# Gráfico de dispersão utilizando os atributos conc e uptake, agrupamento pelo type

#versão antiga da biblioteca
#srn.scatterplot(base.conc, base.uptake, hue = base.Type)
srn.scatterplot(data=base, x='conc', y='uptake', hue='Type')
plt.show()
```

```python
# Seleção de registros específicos da base de dados (Quebec e Mississipi)
q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi']
```

```python
# Subgráfico (1 linha e duas colunas) mostrando gráficos sobre cada região
'''
#versão antiga do seaborn, se a versão abaixo não funcionar,
#remova este comentários (3 aspas simples) e execute este bloco
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(q.conc, q.uptake).set_title('Quebec')
sns.scatterplot(x=q.conc, y=q.uptake)
srn.scatterplot(data=base, x='conc', y='uptake', hue='Type')
plt.subplot(1,2,2)
srn.scatterplot(m.conc, m.uptake).set_title('Mississippi')
plt.tight_layout()
'''

plt.figure(figsize=(10, 5))  # Ajuste o tamanho se necessário
plt.subplot(1, 2, 1)
srn.scatterplot(x=q.conc, y=q.uptake)
plt.title('Quebec')
srn.scatterplot(data=base, x='conc', y='uptake', hue='Type')
plt.subplot(1, 2, 2)
srn.scatterplot(x=m.conc, y=m.uptake)
plt.title('Mississippi')
plt.tight_layout()
plt.show()
```

```python
# Refrigerado e não refrigerado
ch = base.loc[base['Treatment'] == 'chilled']
nc = base.loc[base['Treatment'] == 'nonchilled']
```

## Graficos 3d

```python
# Importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
# Carregamento da base de dados
base = pd.read_csv('orchard.csv')
base
```

```python
# Criação do gráfico 3D, indicando o atributo projection = '3d' e passando três atributos (decrease, rowpos e colpos)
figura = plt.figure()
eixo = figura.add_subplot(1, 1, 1, projection = '3d')
eixo.scatter(base.decrease, base.rowpos, base.colpos)
eixo.set_xlabel('decrease')
eixo.set_ylabel('rowpos')
eixo.set_zlabel('colpos')
```
