# Numpy

```python
import numpy as np
```

```python
# cria uma matriz unidimensional
mt = np.array([12,34,26,18,10])
print(mt)
print(type(mt))
```

```python
#criar o array com um tipo específico
# cria o array como float de 64 bits
mtfloat = np.array([1, 2, 3], dtype = np.float64)
print(mtfloat)
print(type(mtfloat))
mtint = np.array([1, 2, 3], dtype = np.int32)
print(mtint)
print(type(mtint))
```

```python
#mudar o tipo do array
# Podemos transformar tipos de dados de arays
mtnew = np.array([1.4, 3.6, -5.1, 9.42, 4.9999999])
print(mtnew)
#quando transformamos de float para int os valores são rtuncados
mtnewint = mtnew.astype(np.int32)
print(mtnewint)
```

```python
#podemos fazer o inverso também
mt5 = np.array([1,2,3,4])
print(mt5)
mt6 = mt5.astype(float)
print(mt6)
```

```python
#mais de uma dimensão
#cria matriz bidimensional
mt7 = np.array([[7,2,23],[12,8,6],[5,34,23]])
print(mt7)
```

```python
#criar arrays vazios tipificados
#empty significa não são inicializados, não que são vazios
vazio = np.empty([3,2], dtype=int)
print(vazio)
print('------')
#cria uma matriz 4x3 com valores zero
zeros = np.zeros([4,3])
print(zeros)
print('------')
#com valores igual a um
um = np.ones([5,7])
print(um)
print('------')
#cira uma matriz quadrada com diagonal principal com valores 1 e os outros zero
diagonal = np.eye(5)
print(diagonal)
```

```python
#valores aleatórios entre zero e um
ale = np.random.random((5))
print(ale)
print('------')
#valores aleatórios distr, normal contendo negativos
ale2 = np.random.randn((5))
print(ale2)
print('------')
#valores aleatórios 3 x 4
ale3 = (10*np.random.random((3,4)))
print(ale3)
```

```python
#outra forma de gerar aleatórios
#uso de semente
gnr = np.random.default_rng(1)
ale5 = gnr.random(3)
print(ale5)

#gerar inteiros
ale6 = gnr.integers(10, size=(3,4))
print(ale6)
```

```python
#unique remove repetições
j = np.array([11,12,13,14,15,16,17,12,13,18,19,20])
j = np.unique(j)
print(j)
```

```python
#funções específicas
# cria a matriz dimensional k
k = np.array([[7,2,23],[12,8,6],[5,34,23]])
#mostra a matriz k
print(k)
# mostra um elemento especifico da matriz k
print(k[0][1])
# mostra o tamanho das dimensões da matriz k
print(k.shape)
```

```python
#Funções matemáticas
# mostra o maior valor da matriz k
print(k.max())
#mostra o menor valor da matriz k
print(k.min())
#mostra a soma dos valores da matriz k
print(k.sum())
#mostra o valor da média dos valores da matriz k
print(k.mean())
# mostra o valor do desvio padrao (standard derivation) dos valores da matriz k
print(k.std())
```

```python
#Funções universais, aplicadas a todos os elementos
#Mostra o valor da raiz quadrada de todos os elementos
k1 = np.array([1,4,9,16,25,36])
print(np.sqrt(k1))
#Mostra o valor exponencial de todos os elementos
print(np.exp(k1))
```

```python
#extração de elementos
m = np.array([1,2,3,4,5,6])
#Mostra o elemento da posição 2
print(m[1])
print('-------')
#mostra o array criado a partir da posição 0, dois elementos
print(m[0:2])
print('-------')
#Mostra o array criado a partir da 2a posição até todo o restante do array
print(m[1:])
print('-------')
#mostra o array criado a partir da antepenultima posição até o final
print(m[-3:])
```

```python
#extração de linhas e colunas
l = np.array([[4,5],[6,1],[7,4]])
print(l)
print('-------')
#primeira linha, todas as colunas
l_linha_1 = l[0, :]
print(l_linha_1)
print('-------')
#segunda linha
l_linha_2 = l[1, :]
print(l_linha_2)
print('-------')
#todas as linhas, primeira coluna
l_coluna_1 = l[: ,0]
print(l_coluna_1)
print('-------')
#todas as linhas, segunda coluna
l_coluna_2 = l[: ,1]
print(l_coluna_2)
```

```python
#adição e multiplicação de matrizes
n = np.array([[1,2],[3,4]])
o = np.array([[1,1],[1,1]])

res1 = n+o
print(res1)
print('-------')
res2=n*o
print(res2)
print('-------')
p = np.array([[1,2],[3,4],[5,6]])
q = np.array([[2,1]])
print(p+q)
```

```python
#transposição, rearranja um conjunto de 15 elementos de 0 a 14
#em 3 linhas e 5 colunas.
f = np.arange(15).reshape((3,5))
#mostra a matriz transposta entre linha e coluna
print(f)
print("-------")
s = f.T
print(s)
```

```python
# outra forma de fazer, mesmo resultado
r = np.arange(15).reshape((3,5))
print(r)
print("------")
#rearranja um conjunto de 15 elementos
# mostra a matriz transposta entre linha e coluna
s = r.transpose((1,0))
print(s)
```

```python
#expressões lógicas
#usando where
# criando matriz com valores aleatórios positivos e negativos
v = np.random.randn(4,4)
print(v)
#riando mari com valores booleanos baseado no array v
x = (v>0)
print(x)
#criando matriz com valores -1 e 1 baseado nos valores do array x
z = np.where(x > 0, 1, -1)
print(z)
```
